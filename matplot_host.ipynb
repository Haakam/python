```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
std1=np.random.randint(15,45,(100))
std2=np.random.randint(18,40,(100))
```

```python
print(std1)
print(std2)
```

```python
#bins=[15,25,35,45]
```

```python
plt.hist(std1,bins=20,rwidth=0.8)
plt.title("histogram")
plt.xlabel("std age")
plt.ylabel('std catag')
plt.show()
```

```python
plt.hist(std1,bins=20,rwidth=0.8,histtype='stepfilled',color='red')
plt.title("histogram")
plt.xlabel("std age")
plt.ylabel('std catag')
plt.show()
```

```python
plt.figure(figsize=(4,5))# you can change 
plt.hist([std1,std2],histtype='bar',color=['g','r'],label=['std1','std2'])
plt.title("data of std1 and std2",color='m',fontsize='16',fontstyle='italic')
plt.xlabel('data1')
plt.ylabel("y label")
plt.legend()
```

we can chage background of hist

```python
from matplotlib import style
plt.figure(figsize=(4,5))# you can change 
style.use('ggplot')
plt.hist([std1,std2],histtype='bar',color=['g','r'],label=['std1','std2'])
plt.title("data of std1 and std2",color='m',fontsize='16',fontstyle='italic')
plt.xlabel('data1')
plt.ylabel("(y label")
plt.legend()
plt.grid(color='blue',linestyle='-.',linewidth='2')
```
