```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import  pandas as pd

```

```python
from sklearn.datasets import load_breast_cancer
b_cancer=load_breast_cancer()
b_cancer
```

```python
df=pd.DataFrame(np.c_[b_cancer['data'],b_cancer['target']],columns=np.append(b_cancer['feature_names'],['target']))
```

```python
df.head(6)
```

```python
#sns.pairplot(df)
```

```python
sns.pairplot(df,vars=['mean radius','mean texture','mean perimeter','mean area','mean smoothness','mean compactness'],hue='target',palette='Dark2')
```

```python
sns.pairplot(df,x_vars=['mean radius','mean texture','mean perimeter','mean area'], y_vars=['mean smoothness','mean compactness'],hue="target",kind='kde')
```

```python
sns.pairplot(df,x_vars=['mean radius','mean texture','mean perimeter','mean area'], y_vars=['mean smoothness','mean compactness'],hue="target",markers=['*','<'],)
```
