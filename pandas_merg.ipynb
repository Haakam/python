```python
import pandas as pd
df1=pd.DataFrame({'id':[1,2,3,4,5,6],'class':['one','two','three','four','five','six']})
df1
```

```python

df2=pd.DataFrame({'id':[1,2,3,4,5,6],'name':['A','B','C','F','G','W']})
df2
```

```python
pd.merge(df1,df2,on='id')
```

```python
pd.merge(df2,df1,on='id')# on fuction used for index col which has same value on both dara
```

```python

df3=pd.DataFrame({'id':[1,2,3,52,44,22],'name':['A','B','C','F','G','W']})
df3
```

```python
pd.merge(df1,df3,on='id')# on fuction select only those values which have same
```

```python
pd.merge(df1,df3,on='id',how='inner')
```

```python
pd.merge(df3,df2,on='id',how='left')
```

```python
pd.merge(df3,df1,on='id',how= 'outer',indicator=True)
```

```python

df2=pd.DataFrame({'id':[1,2,3,8,5,9],'name':['A','B','C','F','G','W']})
df2
```

```python

df4=pd.DataFrame({'id':[1,2,3,4,5,6],'name':['A','B','C','F','G','W']})
df4
```

```python
pd.merge(df4,df4,on='id',suffixes=('_first','2nd'))
```

```python
pd.merge(df4,df2,on='id',how='inner',suffixes=('_first','2nd'))
```

```python
pd.merge(df4,df2,on='id',how='outer',suffixes=('_first','2nd'))
```

```python
pd.merge(df4,df2,how='inner',right_index=True,left_index=True,suffixes=('_first','2nd'))
```
