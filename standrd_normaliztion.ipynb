# feature scaling 
- normalization or minmax scaler
- standardization or zscore

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
df.columns
```

# feature scaling only apply when your all data in neumerical form so we extract only neuric feature right now

```python
df2=df[['survived', 'pclass','age',  'parch',]]
```

```python
df2
```

```python
df3=df2.fillna(df2.mean())
```

```python
x=df3.drop("survived",axis=1)
```

```python
y=df3["survived"]
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.3,random_state=51,)
```

```python
print("shap ofx test ",x_test.shape)
print("shap of x trian",x_train.shape)
print("shap ofy test ",y_test.shape)
print("shap ofy train ",y_test.shape)
```

# standard scaler learn mean and deviation form x train data

```python
sc=StandardScaler()
sc.fit(x_train)
```

```python
sc.mean_
```

```python
sc.scale_# standard deviation
```

```python
x_train.describe()
```

```python
x_train_sc=sc.transform(x_train)
x_test_sc=sc.transform(x_test)
```

```python
x_train_sc# this data is in two d array we convert it to data fram
```

# for convert data in to data fram

```python
x_train_sc=pd.DataFrame(x_train_sc,columns=['pclass', 'age', 'parch'])
```

```python
x_test_sc=pd.DataFrame(x_test_sc,columns=['pclass', 'age', 'parch'])
```

```python
x_train_sc.head()# standard scaler range is -1 to 0
```

```python
x_train_sc.describe().round(2)
```

```python
mmc=MinMaxScaler()
```

```python
x_train_mm=mmc.fit_transform(x_train)
x_test_mm=mmc.fit_transform(x_test)
```

```python
x_train_mm #this is 2d array now we convert data into datafram
```

```python
x_train_mm=pd.DataFrame(x_train_mm,columns=['pclass', 'age', 'parch'])
x_test_mm=pd.DataFrame(x_test_mm,columns=['pclass', 'age', 'parch'])
```

```python
x_train_mm
```

```python
x_train_mm.describe().round(0)
```
