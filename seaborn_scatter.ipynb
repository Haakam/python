# we use scatter plot in seaborn for maximum use of cols

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("titanic")
df
```

```python
sns.scatterplot(data=df,x="age",y="fare",hue="sex")
```

```python
sns.scatterplot(data=df,x="age",y="fare",hue="sex",style="who",size="who")
```

```python
plt.figure(figsize=(16,9))
sns.scatterplot(data=df,x="age",y="fare",hue="sex",style="who",size="who",sizes=(100,400))
```

```python
sns.scatterplot(data=df,x="who",y="fare",hue="alive",style="alive",size="who",sizes=(100,400),palette="Blues",alpha=1)
```

```python
sns.barplot(data=df,x="age",y="fare",color="red")
sns.lineplot(data=df,x="age",y="fare",color="purple")
sns.scatterplot(data=df,x="age",y="fare",color="green")
```
