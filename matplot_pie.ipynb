```python
import matplotlib.pyplot as plt
```

```python
std=["A","B","C","D"]
marks=[22,34,55,67]
plt.pie([1])
```

```python
std=["A","B","C","D"]
marks=[22,34,55,67]
explode=[0,0.2,0,0.1]
colors=['r','b','black','g']
textprops={"fontsize":14}

plt.pie(marks,labels=std,explode=explode,colors=colors,autopct="%0.2f%%",textprops=textprops,startangle=90, radius=0.9)
plt.legend()
plt.show()
```
