```python
import pandas as pd
```

```python
df1=pd.read_csv("C:\\Users\\Mr\\Desktop\\catdog.csv")
df1
```

```python
df1.pivot_table(index='eyes')#index will give aggregarte value
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='count')#scalar single value
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='sum')
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='count',fill_value='nill')#we can give any value to nan
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='count',fill_value=0)
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='count',dropna=True)
```

```python
df1.pivot_table(index='eyes',columns='mass',aggfunc='sum',dropna=True,margins=True)
```
