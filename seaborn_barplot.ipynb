```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=sns.load_dataset("tips")
df

```

```python
sns.barplot(x=df["day"],y=df.total_bill)
```

```python
sns.barplot(x=df["day"],y=df.total_bill,hue=df.sex)
```

```python
#sns.barplot(x=df["day"],y=df.total_bill,hue=df.sex)both have same result
sns.barplot(x='day',y='total_bill',hue='sex',data=df)
```

```python
order=['Thur','Sun','Sat','Fri']
sns.barplot( data=df,x='day',y='total_bill',hue='sex',order=order)
```

```python
hue_order=['Female','Male']

sns.barplot(x='day',y='total_bill',hue='sex',data=df,order=order,hue_order=hue_order,color="green")
```

```python
hue_order=['Female','Male']

sns.barplot(x='day',y='total_bill',hue='sex',data=df,order=order,hue_order=hue_order,color="green",
           estimator=np.mean)
```

```python


sns.barplot(x='day',y='total_bill',hue='sex',data=df,color="green",
           estimator=np.max)
```

```python


sns.barplot(x='day',y='total_bill',hue='sex',data=df,color="green",
           errorbar='ci',errcolor="purple")
```

```python


sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot")
```

```python


sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot",saturation=0.3)
```

```python


sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot",saturation=0.3,errwidth=5)
```

```python
sns.set()
kwarg={"alpha":0.5,'linestyle':':',"linewidth":5,"edgecolor":'darkblue'}

sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot",saturation=0.3,errwidth=5,**kwarg)
```

```python
sns.set()
sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot",saturation=0.3,errwidth=5,alpha=0.5,linestyle=':',linewidth=5,edgecolor='darkblue')

```

```python
plt.figure(figsize=(16,9))
sns.barplot(x='day',y='total_bill',hue='sex',data=df,
           errorbar='ci',errcolor="purple",palette="hot",saturation=0.3,errwidth=5,alpha=0.5,linestyle=':',linewidth=5,edgecolor='darkblue')
plt.title("sns bar graph between day and toal bill",fontsize=16,color="blue")
plt.xlabel("this is day",fontsize=16,color="red")
plt.ylabel("total bill")
plt.savefig("total bill")
plt.show()
```
