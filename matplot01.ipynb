```python
import matplotlib.pyplot as plt
```

```python
days=[1,2,3,4,5,6,7]
temp=[23.4,45.5,23,67.90,18,40,50]
plt.plot(days,temp,color='b',marker='o',linestyle="--",linewidth="2",markersize="10")
plt.axis([0,8,20,50])
plt.title('this is temperatur of week ')
plt.xlabel('days',loc='center')
plt.ylabel('temp',loc='top')
```

```python
days=[1,2,3,4,5,6,7]
temp=[23.4,45.5,23,67.90,18,40,50]
plt.plot(days,temp,'bo--',linewidth="2",markersize="10")
plt.axis([0,8,20,50])
plt.title('this is temperatur of week ',fontsize=14)
plt.legend(["temp graph"],loc=7)#location has 15 values
plt.xlabel('days',loc='center')
plt.ylabel('temp',loc='top')
```

```python
# use of styles
from matplotlib import style
style.use("ggplot")
days=[1,2,3,4,5,6,7]
temp=[23.4,45.5,23,67.90,18,40,50]
plt.plot(days,temp,'bo--',linewidth="2",markersize="10")
plt.axis([0,8,20,50])
plt.title('this is temperatur of week ',fontsize=14)
plt.legend(["temp graph"],loc=7)#location has 15 values
plt.xlabel('days',loc='center')
plt.ylabel('temp',loc='top')
plt.grid(color='k',linestyle='-.',linewidth=3)
```

```python
# use of styles
from matplotlib import style
style.use("ggplot")
days=[1,2,3,4,5,6,7]
temp=[23.4,45.5,23,67.90,18,40,50]
lhr=[34,5,65,33,54,35,65]
plt.plot(days,temp,'bo--',linewidth="2",markersize="10")
plt.plot(days,lhr,'go--',linewidth="2",markersize="10")
plt.axis([0,8,20,60])
plt.title('this is temperatur of week ',fontsize=14)

plt.xlabel('days',loc='center')
plt.ylabel('temp',loc='top')
plt.legend(["temp","lhr"],loc=7)#location has 15 values
plt.grid(color='k',linestyle='-.',linewidth=3)
```
