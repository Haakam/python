```python
plt.figure(figsize=(9,8))
sns.set()
for i,var in enumerate(mising_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(new_df[var],bins=20,kde_kws={"linewidth":2,'color':'green'})
    plt.show()
```

# when missing values are random we use mean and median when graph is symmentrical we use mean when graph is skewed we use median

```python
df4_mean=new_df.fillna(new_df.mean())# new_df has rows which has null values
df4_mean

```

```python
df4_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(9,8))
sns.set()
for i,var in enumerate(mising_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(new_df[var],bins=20,kde_kws={"linewidth":2,'color':'red'},label="with out fill",color="black")
    sns.distplot(df4_mean[var],bins=20,kde_kws={"linewidth":2,'color':'green'},label="with mean")
    plt.legend()
    plt.show()
```

```python
df5_median=new_df.fillna(new_df.median())# new_df has rows which has null values and fill with median
df5_median.isnull().sum().sum()

```

```python
plt.figure(figsize=(9,8))
sns.set()
for i,var in enumerate(mising_num_var):
    plt.subplot(2,2,1)
    sns.distplot(new_df[var],bins=20,kde_kws={"linewidth":2,'color':'red'},label="with out fill",color="black")
    plt.subplot(2,2,2)
    sns.distplot(df4_mean[var],bins=20,kde_kws={"linewidth":2,'color':'green'},label="with mean")
    plt.subplot(2,2,3)
    sns.distplot(df5_median[var],bins=20,kde_kws={"linewidth":2,'color':'yellow'},label="with median")
    plt.legend()
    plt.show()
```
