```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
days=[1,2,3,4,5,6,7]
temp=[23.4,45.5,23,67.90,18,40,50]
```

```python
df=pd.DataFrame({"days":days,"temp":temp})
```

```python
sns.lineplot(x="days",y="temp",data=df,)
plt.show()
```

```python
df1=sns.load_dataset("tips")
```

```python
df1.columns
```

```python
plt.figure(figsize=(5,7))
sns.set(style='darkgrid')
sns.lineplot(x="total_bill",y="size", data=df1,hue="day",style="day",palette="Accent",
             dashes=False,markers=['o','>'],legend=True)
plt.xlabel("total_bil",fontsize=14,color="red")
plt.ylabel("size",fontsize=14,color="Blue")
plt.title("this is sns chart",color="green")
plt.show()
```

```python
sns.displot(df1["size"],color="red",legend=True,bins=10)
```

```python
sns.displot(df1["size"],color="red",kind="kde",legend=True)
```

```python
sns.displot(df1["total_bill"],color="blue",kind="ecdf",legend=True)
```

```python
sns.displot(df1["total_bill"],color="blue",kind="hist",legend=True,rug=True)
```

```python
sns.distplot(df1["total_bill"],fit=norm)
```

```python
sns.distplot(df1["total_bill"],vertical=True,)
```

```python
sns.displot(df1["total_bill"],label="w",color="green")
plt.title("this is seaborn hist",fontsize="16",color="blue")
plt.xlabel("this is x label",color="red")
plt.ylabel("this is try",color="yellow")
plt.legend()
```

```python
from matplotlib import style
style.use("ggplot")
sns.displot(df1["total_bill"],label="w",color="green",)
plt.title("this is seaborn hist",fontsize="16",color="blue")
plt.xlabel("this is x label",color="red")
plt.ylabel("this is try",color="yellow")
plt.legend()
```
