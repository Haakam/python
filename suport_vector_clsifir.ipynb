# spling data fram into  train and test

```python
x=cancer_df.drop('target',axis=1)
x.head()
```

```python
y=cancer_df['target']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.head()
```

# scaling data

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
```

```python
x_train_sc=sc.fit_transform(x_train)
x_test_sc=sc.fit_transform(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

# The Support Vector Classifier (SVC) is a supervised machine learning algorithm used for classification tasks.

```python
from sklearn.svm import SVC# using raw data
svc_classifier=SVC()
svc_classifier.fit(x_train,y_train)
y_pred_svc=svc_classifier.predict(x_test)
accuracy_score(y_pred_svc,y_test)

```

# logistic regression  an algorithm. It is a statistical method and a machine learning algorithm used for binary classification tasks.

```python
from sklearn.linear_model import LogisticRegression
l_classifier=LogisticRegression()
l_classifier.fit(x_train,y_train)
y_ped_l=l_classifier.predict(x_test)
accuracy_score(y_ped_l,y_test)
```

# he k-nearest neighbors (k-NN) classifier is a simple and intuitive machine learning algorithm used for both classification and regression tasks

```python
from sklearn.neighbors import KNeighborsClassifier
kn_classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
kn_classifier.fit(x_train_sc,y_train)
kn_pred=kn_classifier.predict(x_test_sc)
accuracy_score(kn_pred,y_test)
```

# trained with standard scale data

```python
from sklearn.neighbors import KNeighborsClassifier
kn_classifier=KNeighborsClassifier(n_neighbors=5,metric="minkowski",p=2)
kn_classifier.fit(x_train_sc,y_train)
kn_pred=kn_classifier.predict(x_test_sc)
accuracy_score(kn_pred,y_test)
```

# to implement a Naive Bayes classifier using scikit-learn, you can follow these steps. There are several types of Naive Bayes classifiers available in scikit-learn, such as GaussianNB, MultinomialNB, and BernoulliNB. The choice depends on the nature of your data. Here,  Gaussian Naive Bayes classifier, which is suitable for continuous data.



```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_prd_nb=nb.predict(x_test)
accuracy_score(y_prd_nb,y_test)
```

```python
from sklearn.naive_bayes import GaussianNB # scaled data
nb=GaussianNB()
nb.fit(x_train_sc,y_train)
y_prd_nb=nb.predict(x_test_sc)
accuracy_score(y_prd_nb,y_test)
```

# A Decision Tree classifier is useful for both classification and regression tasks, but here we'll focus on classification.

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(criterion='entropy',random_state=51) # entopy is used when diverse in feature
dt.fit(x_train,y_train)
y_prd_dt=dt.predict(x_test)
accuracy_score(y_prd_dt,y_test)
```

# random forest classifier for large data it can handle missing value too

```python
from sklearn.ensemble import RandomForestClassifier
rc=RandomForestClassifier(n_estimators=20,criterion='entropy',random_state=52)
rc.fit(x_train_sc,y_train)
y_prd_rc=rc.predict(x_test_sc)
accuracy_score(y_prd_rc,y_test)
```

# ada boost classifier(Adaptive Boosting) is a powerful ensemble technique that combines multiple weak classifiers to create a strong classifier

```python
from sklearn.ensemble import AdaBoostClassifier
ada_b=AdaBoostClassifier(DecisionTreeClassifier(criterion="entropy",random_state=52),
                         n_estimators=200,learning_rate=0.1,algorithm='SAMME.R',random_state=1)
ada_b.fit(x_train_sc,y_train)
y_prd_ada=ada_b.predict(x_test_sc)
accuracy_score(y_prd_ada,y_test)
```
